we load the dataset to start preprocessing.


```python
import pandas as pd
df = pd.read_csv("file.csv")

df.head()
```

We check the dataset structure, missing values, duplicate rows, and data types.

```python
df.info()
print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nNumber of duplicate rows:")
print(df.duplicated().sum())
```

These columns do not help in analysis and may affect preprocessing results.

```python
df = df.drop(columns=["Unnamed: 0", "CustomerID", "Transaction_ID"], errors="ignore")
df.head()
```

We fill missing values in numerical columns using the mean.

```python
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print("Missing values after handling:")
print(df.isnull().sum())
```

Duplicate rows are removed to improve data quality.

```python
df = df.drop_duplicates()

print("Number of duplicate rows after removal:")
print(df.duplicated().sum())
```

Detect and handle outliers using the IQR method

```python

numeric_df = df.select_dtypes(include='number')

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

outlier_mask = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))
filtered_rows = ~outlier_mask.any(axis=1)

df = df[filtered_rows]

print("Dataset shape after removing outliers:", df.shape)
```

Only numerical features are used in scaling.

```python
numeric_df = df.select_dtypes(include='number')
numeric_df.head()
```

Normalize numerical features using Min-Max scaling

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(numeric_df), columns=numeric_df.columns)

df_minmax.head()
```

```python
from sklearn.preprocessing import StandardScaler

zscore_scaler = StandardScaler()
df_zscore = pd.DataFrame(zscore_scaler.fit_transform(numeric_df), columns=numeric_df.columns)

df_zscore.head()
```
